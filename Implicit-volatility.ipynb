```python
import pandas as pd
# Load the data 
data = pd.read_csv('microsoft.csv')
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
dates=data["Date"]
closes=data["Adj Close"]
plt.plot(dates[::-1],closes[::-1]) # on parcours de 2012 à 2017
plt.xlabel("Prix options Microsoft")
plt.ylabel("Dates 2012/2017")
plt.title("Cours de marché Microsoft")
```

```python
n=len(dates)
from math import sqrt
volatilites=[]
etape=0
for i in range(60,n,60):
    etape+=i
    s=((data.loc[:etape])["Adj Close"]).std() # écart type des  Adj closes
    v=s/sqrt(60)            # on calcul la volatilité sur une période de deux mois 
    volatilites.append(v)
    
volatilites
```

```python
plt.scatter(range(60,n,60),volatilites[::-1],marker='*') # on parcours les volatilites de 2012 à 2017
plt.xlabel("Dates")
plt.ylabel(" volatilité")
plt.title("courbe de volatilité")
```

On remarque bien que la volatilité diminue quand le prix des options augmente 


Pour avoir une idée sur le prix des options futures, il convient de prendre la volatilité 
correpondant à la plus récente croissance remarquable du prix, dans ce cas 20% pour les mois Mars et Février 2017
